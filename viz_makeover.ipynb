```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly_express as px
from pandas_profiling import ProfileReport
import matplotlib.pyplot as plt
pd.options.display.max_columns = 100
pd.options.display.max_rows = 1000
sns.set_theme()
import plotly.express as px
import chart_studio.plotly as py
```

```python
df = pd.read_csv("rats2.txt", sep="\t")
#df['Created Date'] = pd.to_datetime(df['Created Date'])
df["day"] = df["Created Date"].apply(lambda val: val.split(" ")[0])
df['Created Date'] = pd.to_datetime(df['Created Date'])
df['day'] = pd.to_datetime(df['day'])
df["year"] = df["day"].dt.year
df["month"] = df["day"].dt.month
df = df[df.Status != "Open"]
df = df[df.Status != "Draft"]
df_cnt = df.groupby(["day", "Status"])["Unique Key"].count().reset_index(name="num_rats")
df = pd.merge(df, df_cnt, how="inner", on=["day", "Status"])
df["Location Type"] = df[["Location Type"]].fillna("n/a")
df["Address Type"] = df[["Address Type"]].fillna("n/a")
res_df = df[["day", "num_rats", "Status"]].drop_duplicates().copy()
res_df.head()
res_df.to_csv("viz_comp.csv", sep="\t", index=False, header=True, line_terminator="\n")
```

```python
fig = px.scatter(data_frame = res_df, x="day", y="num_rats", render_mode="auto", color="Status",width=900, height=700,
                 trendline="lowess", labels={
                     "day": "Date",
                     "num_rats": "Rodent Complaints",
                     "Status": "Complaint Status"
                 }, title="NYC Sanitation Department (DSNY) performance for Rodent issues")
tr_line=[]
for  k, trace  in enumerate(fig.data):
    if trace.mode is not None and trace.mode == 'lines':
        tr_line.append(k)

cols = ["#14692a", "#060806", "#f2f5f2"]
for i,id in enumerate(tr_line):
    fig.data[id].update(line_width=3, line_dash="longdashdot", line_color=cols[i])

fig.update_layout(title_x=0.47, title_y=0.89)
fig.show()
```
